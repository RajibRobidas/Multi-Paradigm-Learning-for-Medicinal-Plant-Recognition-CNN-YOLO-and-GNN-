```python
import os
import random
import cv2
import numpy as np
```

```python

# ================= USER INPUT =================
fs = 155        # first missing image number
ls = 200        # last image number

class_prefix = "ramtulsi"  # change per class (e.g., apang, neem, etc.)

src_dir = "/content/drive/My Drive/4_2/ML-4-2/ML-Data/Rare Dataset/Ocimum gatissimum L. (Ram-Tulsi)/"
dst_dir = "/content/drive/My Drive/4_2/ML-4-2/ML-Data/Rare Aug/Ocimum gatissimum L. (Ram-Tulsi)/"
# ==============================================
os.makedirs(dst_dir, exist_ok=True)

```

```python
def augment_image(img):
    h, w = img.shape[:2]

    # 1️⃣ ALWAYS horizontal flip
    img = cv2.flip(img, 1)

    # 2️⃣ ALWAYS rotation (stronger range)
    angle = random.uniform(-35, 35)
    M = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1.0)
    img = cv2.warpAffine(img, M, (w, h), borderMode=cv2.BORDER_REFLECT)

    # 3️⃣ ALWAYS brightness change (stronger & safe)
    value = random.randint(-60, 60)
    img = img.astype(np.int16)
    img = img + value
    img = np.clip(img, 0, 255)
    img = img.astype(np.uint8)

    # Contrast adjustment
    alpha = random.uniform(0.8, 1.3)
    img = np.clip(alpha * img, 0, 255).astype(np.uint8)


    # 4️⃣ ALWAYS zoom (more visible)
    scale = random.uniform(0.8, 1.2)
    M = cv2.getRotationMatrix2D((w // 2, h // 2), 0, scale)
    img = cv2.warpAffine(img, M, (w, h), borderMode=cv2.BORDER_REFLECT)

    return img
```

```python
valid_ext = (".jpg", ".png", ".jpeg")
source_images = [f for f in os.listdir(src_dir) if f.lower().endswith(valid_ext)]

if len(source_images) == 0:
    raise ValueError("❌ No images found in source directory")

print(f"✅ Found {len(source_images)} source images")
```

```python
current_num = fs

while current_num <= ls:
    img_name = random.choice(source_images)
    img_path = os.path.join(src_dir, img_name)

    img = cv2.imread(img_path)
    if img is None:
        continue

    aug_img = augment_image(img)

    save_name = f"{class_prefix}_{current_num}.jpg"
    save_path = os.path.join(dst_dir, save_name)

    cv2.imwrite(save_path, aug_img)
    print(f"Saved: {save_name}")

    current_num += 1

print("🎉 Augmentation completed successfully!")
```

```python
import os
import shutil
import random

# Paths
dataset_path = "/content/drive/My Drive/4_2/ML-4-2/ML-Data/DatasetFinal/Original Dataset/Rare Dataset Original no augmented"
output_path = "/content/drive/My Drive/4_2/ML-4-2/ML-Data/DatasetFinal/Original Dataset/Splitted Original"

train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

random.seed(42)

# Create output folders
for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(output_path, split), exist_ok=True)

# Loop through each class
for class_name in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_name)

    if not os.path.isdir(class_path):
        continue

    images = os.listdir(class_path)
    random.shuffle(images)

    total = len(images)
    train_end = int(total * train_ratio)
    val_end = train_end + int(total * val_ratio)

    train_imgs = images[:train_end]
    val_imgs = images[train_end:val_end]
    test_imgs = images[val_end:]

    for split, split_imgs in zip(
        ["train", "val", "test"],
        [train_imgs, val_imgs, test_imgs]
    ):
        split_class_dir = os.path.join(output_path, split, class_name)
        os.makedirs(split_class_dir, exist_ok=True)

        for img in split_imgs:
            src = os.path.join(class_path, img)
            dst = os.path.join(split_class_dir, img)
            shutil.copy(src, dst)

    print(f"{class_name}: Train={len(train_imgs)}, Val={len(val_imgs)}, Test={len(test_imgs)}")

print("✅ Dataset split completed!")
```
