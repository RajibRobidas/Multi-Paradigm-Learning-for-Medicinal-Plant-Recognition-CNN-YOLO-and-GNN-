🔄 STEP 6 — Convert YOLO → Classification Accuracy

---



```python
def load_true_labels(label_dir):
    y_true = []
    image_files = []

    for file in sorted(os.listdir(label_dir)):
        with open(os.path.join(label_dir, file), "r") as f:
            cls = int(f.readline().split()[0])
            y_true.append(cls)
            image_files.append(file.replace(".txt", ".jpg"))

    return image_files, y_true
```

```python
def yolo_classification_eval(
    model_path,
    image_dir,
    label_dir,
    class_names,
    save_dir
):
    os.makedirs(save_dir, exist_ok=True)
    model = YOLO(model_path)

    image_files, y_true = load_true_labels(label_dir)
    y_pred = []

    for img in image_files:
        result = model(os.path.join(image_dir, img), verbose=False)[0]

        if len(result.boxes) == 0:
            y_pred.append(-1)
        else:
            y_pred.append(int(result.boxes.cls[0]))

    # Remove missed detections
    valid = [i for i,p in enumerate(y_pred) if p != -1]
    y_true = [y_true[i] for i in valid]
    y_pred = [y_pred[i] for i in valid]

    # Accuracy
    acc = accuracy_score(y_true, y_pred)
    print("Classification Accuracy:", acc)

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="viridis",
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("YOLO Classification Confusion Matrix")
    plt.savefig(os.path.join(save_dir, "confusion_matrix_cls.png"), dpi=300)
    plt.show()

    # Classification Report
    report = classification_report(
        y_true, y_pred, target_names=class_names, output_dict=True
    )
    pd.DataFrame(report).transpose().to_csv(
        os.path.join(save_dir, "classification_report_cls.csv")
    )

    # Per-class Accuracy
    per_class_acc = np.diag(cm) / cm.sum(axis=1)
    plt.figure(figsize=(10,5))
    plt.bar(class_names, per_class_acc)
    plt.xticks(rotation=45)
    plt.ylabel("Accuracy")
    plt.title("YOLO Per-Class Accuracy")
    plt.savefig(os.path.join(save_dir, "per_class_accuracy_cls.png"), dpi=300)
    plt.show()

    return acc
```
